```python
import pandas_datareader.data as pdr
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
from datetime import datetime
%matplotlib inline
```

```python
start = datetime(2021, 6, 1)
end = datetime(2022, 1, 1)
```

```python
BTC = pdr.DataReader('BTC-USD','yahoo',start,end)
ETH = pdr.DataReader('ETH-USD','yahoo',start,end)
DOGE = pdr.DataReader('DOGE-USD','yahoo',start,end)
SHIB = pdr.DataReader('SHIB-USD','yahoo',start,end)
```

```python
BTC
```

```python
tickers = ['BTC-USD','ETH-USD','DOGE-USD','SHIB-USD']
```

```python
crypto = pd.concat([BTC,ETH,DOGE,SHIB],axis=1,keys=tickers)
crypto.head()
```

```python
crypto.columns.names = ['Crypto Ticker','Crypto Info']
```

```python
crypto.head()
```

```python
crypto.xs(key='Close',axis=1,level='Crypto Info').max()
```

```python
crypto.xs(key='Close',axis=1,level='Crypto Info').min()
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick+' Return'] = crypto[tick]['Close'].pct_change()
```

```python
returns.head()
```

```python
sns.pairplot(returns[1:])
```

```python
returns.idxmin()
```

```python
returns.idxmax()
```

```python
returns.std()
```

```python
sns.displot(returns['SHIB-USD Return'],color='orange',bins=20)
```

```python
sns.displot(returns['BTC-USD Return'],color='yellow',bins=20)
```

```python
for tick in tickers:
    crypto[tick]['Close'].plot(label=tick,figsize=(12,4))
plt.legend()
```
